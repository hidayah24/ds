```python
df3 = df2.drop(df2.loc[[0,8,11,14,18,23,41,42,43,56]].index)
df3['Fund allotted(in ₹crores)'].sum()
```

```python
df4 = df2.iloc[[0,8,11,14,18,23,41,42,43],:]
row = {'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)': df3['Fund allotted(in ₹crores)'].sum()}
df4 = df4.append(row, ignore_index = True)
df4
```

```python
df4.plot.bar(x='Department /Ministry', y='Fund allotted(in ₹crores)')
```

```python
import matplotlib.pyplot as plt
```

```python
df5 = df4["Fund allotted(in ₹crores)"]
labels = df4["Department /Ministry"]
plt.figure(figsize=(7,7))
plt.pie(df5, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution Financial Budget in India", fontsize=20)
plt.show()
```
